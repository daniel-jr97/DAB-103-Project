```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset 
unemployment = pd.read_csv('C:/Users/felix/Desktop/unmp.csv',skipfooter=26,engine='python')
unemployment
```

```python
#Assigning to a dataframe variable.
unmp=unemployment
```

```python
#Dropping blank columns 
unmp=unmp.drop('Unnamed: 7',axis=1)
```

```python
unmp
```

```python
#Renaming the columns for easy of access. 
unmp.rename(columns = {'North American Industry Classification System (NAICS) 4':'Industry'}, inplace = True)
```

```python
unmp
```

```python
#Converting objects to float to plot and calculate. 
unmp['2016'] = unmp['2016'].astype(float)
unmp['2017'] = unmp['2017'].astype(float)
unmp['2018'] = unmp['2018'].astype(float)
unmp['2019'] = unmp['2019'].astype(float)
unmp['2020'] = unmp['2020'].astype(float)
unmp['2021'] = unmp['2021'].astype(float)
```

```python
#Plots showing the distribution of unemployment for each year
unmp.plot.scatter(x = "2016", y = "Industry", ylabel= "Industry", xlabel = "Percentage of unemployed people", title = "Unemployment percentage in 2016" )
unmp.plot.scatter(x = "2017", y = "Industry", ylabel= "Industry", xlabel = "Percentage of unemployed people", title = "Unemployment percentage in 2017" )
unmp.plot.scatter(x = "2018", y = "Industry", ylabel= "Industry", xlabel = "Percentage of unemployed people", title = "Unemployment percentage in 2018" )
unmp.plot.scatter(x = "2019", y = "Industry", ylabel= "Industry", xlabel = "Percentage of unemployed people", title = "Unemployment percentage in 2019" )
unmp.plot.scatter(x = "2020", y = "Industry", ylabel= "Industry", xlabel = "Percentage of unemployed people", title = "Unemployment percentage in 2020" )
unmp.plot.scatter(x = "2021", y = "Industry", ylabel= "Industry", xlabel = "Percentage of unemployed people", title = "Unemployment percentage in 2021" )
```

```python
#Bar graph showing the comparision of unemployment across industries over the time span. 
unmp.plot(kind="bar", x="Industry", y=["2016", "2017", "2018","2019","2020","2021"])
```

```python
#Plot comparing 2016 and 2017 unemployment rate gap according to industry 
plt.xticks(rotation=90)
g_2016=sns.lineplot(x=unmp["Industry"],y=unmp["2016"])
g_2017=sns.lineplot(x=unmp["Industry"],y=unmp["2017"])
```

```python
#Plot comparing 2018 and 2019 unemployment rate gap according to industry 
plt.xticks(rotation=90)
g_2018=sns.lineplot(x=unmp["Industry"],y=unmp["2018"])
g_2019=sns.lineplot(x=unmp["Industry"],y=unmp["2019"])
```

```python
#Plot comparing 2020 and 2021 unemployment rate gap according to industry  
plt.xticks(rotation=90)
g_2020=sns.lineplot(x=unmp["Industry"],y=unmp["2020"])
g_2021=sns.lineplot(x=unmp["Industry"],y=unmp["2021"])
```
